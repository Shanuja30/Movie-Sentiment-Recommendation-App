```python
data = data1.merge(data2, on='title')
data.head()
```

```python
data = data.drop(columns = ['budget', 'revenue' , 'homepage' , 'id' , 'movie_id' , 'original_title' , 'popularity', 'vote_average', 'vote_count', 'production_countries', 'release_date', 'runtime', 'status', 'tagline', 'spoken_languages'])
```

```python
data = data.dropna(subset=['overview']).reset_index(drop=True)
```

```python
data.isna().sum()
```

```python
# Step 1: Process columns

# 1. Function to extract director from crew
def get_director(crew_list):
    try:
        for member in eval(crew_list):
            if member['job'] == 'Director':
                return member['name']
    except:
        return ''
    return ''

# 2. Function to take top 3 actors from cast
def get_top_cast(cast_list):
    try:
        actors = [member['name'] for member in eval(cast_list)]
        return actors[:3]
    except:
        return []

# 3. Function to extract names from stringified lists (genres, keywords, production_companies)
def get_names(column):
    try:
        return [item['name'] for item in eval(column)]
    except:
        return []

# Apply functions
data['director'] = data['crew'].apply(get_director)
data['cast'] = data['cast'].apply(get_top_cast)
data['genres'] = data['genres'].apply(get_names)
data['keywords'] = data['keywords'].apply(get_names)
data['production_companies'] = data['production_companies'].apply(get_names)

# Check the processed data
data[['title', 'director', 'cast', 'genres', 'keywords', 'production_companies']].head()
```

```python
# Step 2: Combine into a single tags column

# Convert lists to strings and combine all features into one column
data['tags'] = data['overview'] + ' ' + \
               data['genres'].apply(lambda x: ' '.join(x)) + ' ' + \
               data['keywords'].apply(lambda x: ' '.join(x)) + ' ' + \
               data['cast'].apply(lambda x: ' '.join(x)) + ' ' + \
               data['director']

# Check the first few entries of tags
data[['title', 'tags']].head()
```

```python
data.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Fit and transform tags column
vectors = tfidf.fit_transform(data['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute similarity matrix
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie_name, data=data, similarity=similarity):
    # Find the index of the movie
    movie_index = data[data['title'] == movie_name].index[0]
    
    # Get similarity scores for this movie with all other movies
    distances = list(enumerate(similarity[movie_index]))
    
    # Sort movies based on similarity scores (highest first)
    movies_list = sorted(distances, key=lambda x: x[1], reverse=True)
    
    # Skip the first one (the movie itself) and take top 5
    recommended_movies = []
    for i in movies_list[1:6]:
        recommended_movies.append(data.iloc[i[0]].title)
    
    return recommended_movies
```

```python
recommend("The Dark Knight")
```

```python
import joblib

joblib.dump(tfidf, 'tfidf_vectorizer_Rec.pkl')

joblib.dump(vectors, 'vectors_Rec.pkl')

data[['title', 'tags']].to_csv('movie_data_processed.csv', index=False)
```
